```python
# Libraries #

import os
from dotenv import load_dotenv
from openai import OpenAI
from pypdf import PdfReader

load_dotenv(override=True)
```

```python
# Keys

openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key is None:
    raise ValueError("OPENAI_API_KEY is not set")
else:
    print("OPENAI_API_KEY is set")
```

```python
# Get the CV/Resume

reader = PdfReader("my_cv.pdf")
cv_text = ""
for page in reader.pages:
    cv_text += page.extract_text() + "\n"

print(cv_text)
```

```python
# System Prompt

system_prompt = f"""
You are my assistant. You are responsible to reply to questions about my CV/Resume and my career.
You are my cv here:\n{cv_text}. Use it to reply to the user's question.
"""
print(system_prompt)
```

```python
# user_prompt = "Generate a content for the page."

# # Either a system role or a user role
# messages = [
#     {"role": "system", "content": system_prompt},
#     {"role": "user", "content": user_prompt},
# ]

model_name = "gpt-4o-mini"
```

```python
client = OpenAI()



# user_prompt = input("Enter your question: ")

# if user_prompt == "exit":
#     break

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "What is your name and uni"},
]

response = client.chat.completions.create(
    model=model_name,
    messages=messages,
)

# response: the full API response object returned by client.chat.completions.create()
# response.choices: a list of possible completions (the API can return more than one choice)
# response.choices[0]: the first completion (index 0 = first item in the list)
# response.choices[0].message: the structured message object (with role + content fields)
# response.choices[0].message.content: the actual text string generated by the model

response_text = response.choices[0].message.content

print(response_text)
```
