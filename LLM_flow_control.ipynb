```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import json
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
```

```python
def multiplication(x, y):
    result = x * y
    print(result)

def add(x, y):
    result = x + y
    print(result)

def subtraction(x, y):
    result = x - y
    print(result)

def devide(x, y):
    result = x / y
    print(result)
```

```python
system_prompt = "You are a calculater router.\n"
system_prompt += "Tasks: Your responsibility is to take the user prompt and decide his intent.\n"
system_prompt += "You have 4 routes. [*,+,-,/]\n"
system_prompt += "You extract two numbers. [x,y].\n"
system_prompt += "For example. User Prompt: add 2 to 3. route = +, x = 2, y = 3\n"
system_prompt += "For example. User Prompt: 3 * 5. route = *, x = 3, y = 5\n"
system_prompt += "Output JSON: {\"route\":\"+\", \"x\":x, \"y\":y}\n"

model = "gpt-4o-mini"
```

```python
# system_prompt += "You have 4 routes. [multiplication,add,subtraction,devide]\n"
# system_prompt += "For example. User Prompt: add 2 to 3. route = add, x = 2, y = 3\n"
# system_prompt += "For example. User Prompt: 3 * 5. route = multiplication, x = 3, y = 5\n"
```

```python
def router(request):
    client = OpenAI()
    response = client.chat.completions.create(
    model=model,
    messages=[{"role": "system", "content": system_prompt},
            {"role": "user", "content":request}],
    response_format={"type": "json_object"}
    )

    answer = response.choices[0].message.content
    answer = json.loads(answer)
    route = answer["route"]
    x = int(answer["x"])
    y = int(answer["y"])
 
    print(answer)
    
    if route == "*":
        multiplication(x,y)

    elif route == "+":
        add(x,y)

    elif route == '-':
        subtraction(x, y)
    
    else:
        devide(x,y)

 
router("7 * 5")
```
